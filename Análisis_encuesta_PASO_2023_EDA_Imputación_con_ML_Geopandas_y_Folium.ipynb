# 9. EDAD Y GENERO

#### Reemplazo de valores faltantes de Género y Edad

Para reemplazar valores faltantes de:
* Edad: Se hará un filtro por provincia y luego por nivel educativo. Se calculará el promedio de la edad de cada nivel educativo por provincia y se imputará esos valores.

* Genero: Se hará un filtro por provincia, nivel educativo y voto en las PASO por ESPACIO, dado que se observa que el voto por espacio, mayormente no cambia drásticamente de una elección a otra. En base a eso se llevará a cabo un modelo de clasificación con el algoritmo random forest para estimar los valores faltantes de género.

#### EDAD

```python
valores_faltantes = df['Edad'].isna().sum()
print(f"Total de valores faltantes en la columna 'Edad': {valores_faltantes}")
```

```python
#Vamos a modificar la imputación de la columna Nivel Educativo. Y vamos a poner, por ejemplo, en vez de Secundario completo o incompleto, simplemente "Secundario"
#Antes, voy a ver los valores único de la columna Nivel educativo.

df['Nivel educativo'].unique()
```

```python
mapeo={
    "Secundario incompleto/completo": "Secundario",
    "Universitario o posgrado incompleto/completo o más":"Universitario",
    "Primario incompleto/completo":"Primario",
    "Terciario incompleto/completo":"Terciario"
    }

df['Nivel educativo'] = df['Nivel educativo'].replace(mapeo)

print(df['Nivel educativo'].value_counts())
```

El siguiente código a utilizar realiza el siguiente procedimiento:

* Calcula el promedio de edad para cada combinación de provincia y nivel educativo y almacena esos promedios en promedios_por_provincia_nivel.

* Luego, recorre todas las filas del DataFrame original df y verifica si la columna 'Edad' tiene un valor faltante (NaN).

* Si encuentra un valor faltante, extrae la provincia y el nivel educativo correspondientes a esa fila.

* Luego, busca en promedios_por_provincia_nivel si existe un promedio de edad asociado a esa combinación de provincia y nivel educativo.

* Si se encuentra un promedio, se asigna ese valor como el nuevo valor de 'Edad' en la fila, lo que imputa el valor faltante.

```python
promedios_por_provincia_nivel = df.groupby(['Provincia', 'Nivel educativo'])['Edad'].mean()

for index, row in df.iterrows():
    if pd.isna(row['Edad']):
        provincia = row['Provincia']
        nivel_educativo = row['Nivel educativo']
        if (provincia, nivel_educativo) in promedios_por_provincia_nivel:
            edad_promedio = promedios_por_provincia_nivel[(provincia, nivel_educativo)]
            df.at[index, 'Edad'] = edad_promedio
```

```python
#Aca creo un dataframe reducido con las columnas que me interesan para predecir.
df_reducido= df_sina[['Genero', 'Provincia', 'Nivel educativo', 'Espacio']]
```

```python
#Aca lo que voy a hacer es un modelo de Clasificación con Random Forest para imputar los valores de Género faltantes.
#La estimación de los valores se hará de acuerdo a la Provincia, Nivel educativo y Espacio al que votó en las PASO.
import pandas as pd
from sklearn.ensemble import RandomForestClassifier

# Codifico variables categóricas
df_encoded = pd.get_dummies(df_reducido, columns=['Provincia', 'Nivel educativo', 'Espacio'], drop_first=True)

# Divido en conjunto de entrenamiento y prueba
train_data = df_encoded[df_encoded['Genero'] != '']
test_data = df_encoded[df_encoded['Genero'] == '']

features = train_data.drop(columns=['Genero'])
target = train_data['Genero']

forest = RandomForestClassifier(n_estimators=100, random_state=42)

forest.fit(features, target)

# Realizo predicciones en el conjunto de prueba
test_features = test_data.drop(columns=['Genero'])
predictions = forest.predict(test_features)

# Asigno las predicciones como los valores imputados
df_reducido.loc[df_reducido['Genero'] == '', 'Genero'] = predictions


df_reducido
```

# 12. Geopandas y Folium





#### Encuestados por provincia

```python
import pandas as pd
import geopandas as gpd
import folium

provincia_counts = df_final['Provincia'].value_counts()

# Calcular el porcentaje en relación al total
total_personas = len(df_final)
porcentaje_por_provincia = (provincia_counts / total_personas) * 100

encuestados_provincias= porcentaje_por_provincia.reset_index()
encuestados_provincias.columns = ['nombre', 'Porcentaje']
```

```python

encuestados_provincias['nombre'] = encuestados_provincias['nombre'].str.strip()

# Lee el archivo GeoJSON con los límites de las provincias argentinas desde el enlace
# https://www.kaggle.com/datasets/pablomgomez21/geojson-file-provincias-argentinas/ de aca se descarga el archivo

provincias_geojson = '/content/drive/My Drive/ProvinciasArgentina.geojson'

provincias_gdf = gpd.read_file(provincias_geojson)

provincias_gdf['nombre'] = provincias_gdf['nombre'].replace('Capital Federal', 'Ciudad Autónoma de Buenos Aires')
provincias_gdf['nombre'] = provincias_gdf['nombre'].replace('Tierra del Fuego', 'Tierra del Fuego, Antártida e Islas del Atlántico Sur')
```

```python

provincias_combinado_encuestados = provincias_gdf.merge(encuestados_provincias, on='nombre', how='left')

provincias_combinado_encuestados= provincias_combinado_encuestados[['nombre', 'Porcentaje', 'geometry']]
```

```python
provincias_combinado_encuestados
```

```python
import folium

argentina_coords = [-34.611778, -58.417308]

mapa_encuestados = folium.Map(location=argentina_coords, zoom_start=5)

folium.Choropleth(
    geo_data=provincias_combinado_encuestados,
    name='choropleth',
    data=provincias_combinado_encuestados,
    columns=['nombre', 'Porcentaje'],
    key_on='feature.properties.nombre',
    fill_color='Greens',
    fill_opacity=0.8,
    line_opacity=0.4,
    legend_name='Porcentaje de votos por Provincia',
    position='bottomright',
    threshold_scale=[0, 1, 3, 6, 10 , 25, 45]
).add_to(mapa_encuestados)

mapa_encuestados
mapa_encuestados.save('mapa_encuestados.html')
```
